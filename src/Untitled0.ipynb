```python
import subprocess
import time

print("Starting Streamlit app with localtunnel...")

# Start localtunnel and capture its output for the URL
st_tunnel = subprocess.Popen('npx localtunnel --port 8501', shell=True, stdout=subprocess.PIPE)
time.sleep(4) # Give localtunnel some time to establish the connection and print the URL
streamlit_link = st_tunnel.stdout.readline().decode('utf-8').strip()
print(f"Streamlit Link: {streamlit_link}")

# Start the Streamlit app in the background
subprocess.Popen("streamlit run app/Home.py --server.port 8501", shell=True)

print("\n✅ Streamlit app and localtunnel should be running.")
print("Access your Streamlit app at the link above.")
```

```python
print("Attempting to stop Streamlit process...")
# Find the process ID (PID) of the Streamlit app
# We look for processes containing 'streamlit run' and 'app.py'
# Using pgrep to find the PID more robustly
try:
    # Find the PID of the streamlit process
    pid_command = !pgrep -f 'streamlit run app.py'
    if pid_command:
        pid = pid_command[0].strip()
        print(f"Found Streamlit process with PID: {pid}")
        # Kill the process
        !kill {pid}
        print("Streamlit process stopped.")
    else:
        print("No Streamlit process found running 'app.py'.")
except Exception as e:
    print(f"An error occurred while trying to stop Streamlit: {e}")
```
